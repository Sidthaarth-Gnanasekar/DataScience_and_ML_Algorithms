```python
import pandas as pd
df = pd.read_csv('titanic.csv')
df
```

```python
df1=df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns')
df1
```

```python
df1.dropna(how='any', inplace=True)
df1
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
df1['Gender']=le.fit_transform(df1.Sex)
```

```python
df2=df1.drop(['Sex','Survived'],axis='columns')
df2
```

```python
from sklearn import tree
model=tree.DecisionTreeClassifier()
```

```python
from sklearn.model_selection import train_test_split 
x_train, x_test, y_train, y_test=train_test_split(df2,df1.Survived,test_size=0.2)
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
model.predict([[2,25,50,1]])
```
